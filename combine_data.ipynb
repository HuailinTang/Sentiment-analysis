```python
import pandas as pd  
import itertools
from google.colab import drive
import os
```

```python
drive.mount('/content/drive', force_remount=True)
```

```python
results_df = pd.DataFrame(columns=['created_at', 'text'])
```

```python
folder_path = './drive/MyDrive/Comp 642 data/'
batch_size = 10000 # number of rows to read in each batch

for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        file_path = os.path.join(folder_path, filename)
        print(file_path)

        # Iterate over the file in chunks and process each chunk
        json_reader = pd.read_json(file_path, lines=True, chunksize=batch_size)
        for chunk in json_reader:
            processed_chunk = chunk[['created_at', 'text']]
            results_df = pd.concat([results_df, processed_chunk])

print(f"Processed all files and stored the results in a single dataframe with {results_df.shape[0]} rows")
```

```python
# Write the results to a CSV file
results_df.to_csv("./drive/MyDrive/Comp 642 data/tweets_preprocessed.csv", index=False)
```
